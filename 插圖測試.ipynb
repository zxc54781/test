```python
import matplotlib.pyplot as plt
import numpy as np
#import matplotlib.image as mpimg # mpimg 用于读取图片


#from matplotlib.patches import Circle
from matplotlib.offsetbox import (TextArea, DrawingArea, OffsetImage,
                                  AnnotationBbox)
from matplotlib.cbook import get_sample_data
```

```python
def draw1(p1_rk , p2_rk,p1_rp , p2_rp,p1_wk, p2_wk,p1_wp , p2_wp):


    ## 以下為控制的座標 rk =紅王
    c=0.095 # x constance
    d=0.085 #y constance
    x_rk=0.1157*p1_rk+c
    y_rk=0.1174*p2_rk+d
    x_rp=0.1157*p1_rp+c
    y_rp=0.1174*p2_rp+d
    x_wk=0.1157*p1_wk+c
    y_wk=0.1174*p2_wk+d
    x_wp=0.1157*p1_wp+c
    y_wp=0.1174*p2_wp+d


    fig, ax = plt.subplots(nrows=1, ncols=1, sharex=True, sharey=True,figsize=(5,5),dpi=80)
    #imgP=plt.imread('wood.png')

    Fn_Wo = get_sample_data("wood.png", asfileobj=False)
    Fn_RK = get_sample_data("red_king1.png", asfileobj=False)
    Fn_RP = get_sample_data("red_1.png", asfileobj=False)
    Fn_WK = get_sample_data("white_king1.png", asfileobj=False)
    Fn_WP = get_sample_data("white_1.png", asfileobj=False)

    Wo_img = plt.imread(Fn_Wo, format='png')
    RK_img = plt.imread(Fn_RK, format='png')
    RP_img = plt.imread(Fn_RP, format='png')
    WK_img = plt.imread(Fn_WK, format='png')
    WP_img = plt.imread(Fn_WP, format='png')

    Woimagebox = OffsetImage(Wo_img, zoom=1)
    RKimagebox = OffsetImage(RK_img, zoom=0.14)
    RPimagebox = OffsetImage(RP_img, zoom=0.14)
    WKimagebox = OffsetImage(WK_img, zoom=0.14)
    WPimagebox = OffsetImage(WP_img, zoom=0.14)
    #xy=[0.5,0.5]

    Wo = AnnotationBbox(Woimagebox, [0.5,0.5],
                        xybox=(0,0),
                        xycoords='data',
                        boxcoords="offset points",
                        pad=1,
                        frameon=False
                             )

    RK = AnnotationBbox(RKimagebox, [x_rk,y_rk],
                        xybox=(0,0),
                        xycoords='data',
                        boxcoords="offset points",
                        pad=1,
                        frameon=False
                             )
    c=0.095
    d=0.085+0.1174*3
    RP = AnnotationBbox(RPimagebox, [x_rp,y_rp],
                        xybox=(0, 0),
                        xycoords='data',
                        boxcoords="offset points",
                        pad=1,
                        frameon=False
                             )
 
    WK = AnnotationBbox(WKimagebox, [x_wk,y_wk],
                        xybox=(0, 0),
                        xycoords='data',
                        boxcoords="offset points",
                        pad=1,
                        frameon=False
                             )

    WP = AnnotationBbox(WPimagebox, [x_wp,y_wp],
                        xybox=(0, 0),
                        xycoords='data',
                        boxcoords="offset points",
                        pad=1,
                        frameon=False
                             )
    #ax.imshow(imgP)
    ax.add_artist(Wo)
    ax.add_artist(RK)
    ax.add_artist(RP)
    ax.add_artist(WK)
    ax.add_artist(WP)
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)

    plt.axis('off')
    plt.show()
```
